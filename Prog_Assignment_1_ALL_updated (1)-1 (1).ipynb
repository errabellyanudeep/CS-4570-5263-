a)Create and draw a 6-qubit/6-classical bit quantum circuit where you apply an X followed by H gate to each qubit using a for loop and measure all the bits.


```python
# Define quantum circuit
quantum_circuit = QuantumCircuit(6,6)

#gates 

# Apply X gate followed by H gate to each qubit using a for loop
for qubit in range(6):
    quantum_circuit.x(qubit)  # Applicat X gate
    quantum_circuit.h(qubit)  # Apply H gate


#measure
quantum_circuit.measure(range(6), range(6))

#draw
quantum_circuit.draw()
```

b) Create and draw a 5-qubit/5-classical bit quantum circuit where you apply the gates listed below using a loop and measure.

H gate on qubit 0

CX with qubit 0 as the control and qubit 1 as the target

H gate on qubit 1

CX with qubit 1 as the control and qubit 2 as the target

H gate on qubit 2

CX with qubit 2 as the control and qubit 3 as the target

```python
# write your code here    
quantum_circuit = QuantumCircuit(5,5)

# gates
for  gate  in range(5):
    if gate == 0:
        quantum_circuit.h(gate)
        quantum_circuit.cx(gate,gate+1)
        
    elif gate == 1:
        quantum_circuit.h(gate)
        quantum_circuit.cx(gate,gate+1)

    elif gate == 2:
        quantum_circuit.h(gate)
        quantum_circuit.cx(gate,gate+1)
    
# Measure the state
for i in range(5):
    quantum_circuit.measure(i, i)
        

# Draw
quantum_circuit.draw()
```
